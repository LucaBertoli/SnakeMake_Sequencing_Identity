```python
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np

results = {}

for tech_name, tech_df in tech_dfs.items():

    df = tech_df.copy()
    df["Q"] = -10 * np.log10(1 - df["identity_with_ins_filtered"])
    df = df.replace([np.inf, -np.inf], np.nan).dropna(subset=["Q"])

    # Modello con interazione
    model = smf.ols("Q ~ Cycle * C(BQ_bin)", data=df).fit()

    # Trova i bin
    bins = sorted(df["BQ_bin"].unique(), key=str)
    ref_bin = bins[0]

    tech_slopes = {}
    tech_pvalues = {}

    # --- Slope + pvalue per il bin di riferimento ---
    base_slope = model.params["Cycle"]
    base_pvalue = model.pvalues["Cycle"]

    tech_slopes[ref_bin] = base_slope
    tech_pvalues[ref_bin] = base_pvalue

    # --- Slopes e pvalue per gli altri bin ---
    for b in bins[1:]:
        term = f"Cycle:C(BQ_bin)[T.{b}]"

        slope = base_slope + model.params.get(term, 0)
        pval = model.pvalues.get(term, np.nan)

        tech_slopes[b] = slope
        tech_pvalues[b] = pval

    # Interpretazione qualitativa
    interpretation = {
        b: (
            "Crescente" if s > 0.01
            else "Decrescente" if s < -0.01
            else "Stabile"
        )
        for b, s in tech_slopes.items()
    }

    # Costruzione tabella
    results[tech_name] = pd.DataFrame({
        "Slope": tech_slopes,
        "p-value": tech_pvalues,
        "Interpretazione": interpretation
    })

# --- Output ---
for tech, df_res in results.items():
    print("\n###", tech)
    print(df_res)
```
