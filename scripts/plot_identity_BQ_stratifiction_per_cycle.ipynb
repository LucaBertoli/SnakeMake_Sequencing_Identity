```python
tech_dfs = {
    "Illumina NovaSeq X": df_tot_ns,
    "Element AVITI": df_tot_elm,
    "GeneMind SurfSeq 5000": df_tot_gm,
    "MGI T1": df_tot_mgi
}

tech_dfs_1 = {
    "Illumina NovaSeq X": df_tot_noBin_ns_R1,
    "Element AVITI": df_tot_noBin_elm_R1,
    "GeneMind SurfSeq 5000": df_tot_noBin_gm_R1,
    "MGI T1": df_tot_noBin_mgi_R1
}

tech_dfs_2 = {
    "Illumina NovaSeq X": df_tot_noBin_ns_R2,
    "Element AVITI": df_tot_noBin_elm_R2,
    "GeneMind SurfSeq 5000": df_tot_noBin_gm_R2,
    "MGI T1": df_tot_noBin_mgi_R2
}

tutti_tech_dfs = [tech_dfs, tech_dfs_1, tech_dfs_2]

for i, dfs in enumerate(tutti_tech_dfs):

    if i == 0:
        mode = ""
    elif i == 1:
        mode = " Read 1"
    else:
        mode = " Read 2"

    print(f"Plot gruppo {i}{mode}")

    # --- CALCOLA QUALITÀ MEDIA PER CICLO ---
    tech_cycle_quality = {}

    for tech_name, df in dfs.items():

        # Evita di modificare il dataframe originale
        df_local = df.copy()

        # Calcola totale basi
        df_local["total_bases"] = df_local["match"] + df_local["mismatch"] + df_local["insertion"]

        # Raggruppa per ciclo e calcola la media pesata del Qscore
        grouped = (
            df_local.groupby("Cycle", observed=False)
                .apply(lambda x: (
                    -10 * np.log10(
                        np.average(10 ** (-x["BaseQuality"] / 10), weights=x["total_bases"])
                    )
                    if x["total_bases"].sum() > 0 else np.nan
                ))
                .reset_index(name="mean_BaseQuality")
        )

        tech_cycle_quality[tech_name] = grouped

    # --- PLOT ---
    fig, ax = plt.subplots(figsize=(8, 5))

    for j, (tech_name, grouped) in enumerate(tech_cycle_quality.items()):
        ax.plot(
            grouped["Cycle"],
            grouped["mean_BaseQuality"],
            label=tech_name,
            color=colori_matplotlib[j % len(colori_matplotlib)],
            linewidth=0.8,
            marker='o',
            markersize=2,
            alpha=0.8
        )

    # Titolo
    ax.set_title(
        f"Sequencer-assigned Qscore per Cycle{mode} (logarithm of mean error probability)"
    )

    # Assi
    ax.set_xlabel("Cycle")
    ax.set_ylabel("Mean Qscore")
    ax.grid(True, alpha=0.3)
    ax.set_ylim(0, 43)

    # Legenda globale sotto il grafico
    fig.legend(
        tech_cycle_quality.keys(),
        loc="lower center",
        ncol=len(tech_cycle_quality),
        fontsize=10,
        frameon=True
    )

    plt.tight_layout(rect=[0, 0.05, 1, 1])
    plt.show()

```

```python
####normalizzazione per base
# Dizionario con i dataframe e il nome della tecnologia
tech_dfs = {
    "Illumina NovaSeq X": df_ns,
    "Element AVITI": df_elm,
    "GeneMind SurfSeq 5000": df_gm,
    "MGI T1": df_mgi
}
 
# Trova tutti i BQ_bin disponibili (assumendo che siano simili tra i dataframe)
all_bins = sorted(
    set().union(*[df["BQ_bin"].unique() for df in tech_dfs.values()]),
    key=lambda x: str(x)
)

# --- FIGURA CON DUE RIGHE DI SUBPLOT (3 col x 2 righe) ---
fig, axes = plt.subplots(2, 3, figsize=(15, 8), sharex='col')
axes_qscore = axes[0]
axes_bases = axes[1]

for i, quality in enumerate(all_bins[:3]):  # usa i primi 3 bin
    ax_q = axes_qscore[i]
    ax_b = axes_bases[i]

    for j, (tech_name, tech_df) in enumerate(tech_dfs.items()):
        sub = tech_df[tech_df["BQ_bin"] == quality].sort_values("Cycle").copy()
        if sub.empty:
            continue

        # Calcolo Qscore (basato su identità)
        qscore = -10 * np.log10(1 - sub["identity_with_ins_filtered"])

        # Numero di basi per ciclo per questo bin
        sub["num_bases"] = sub["match"] + sub["mismatch"] + sub["insertion"]

        # ✅ Normalizza per la somma totale di basi di *quel ciclo* nella stessa tecnologia
        # Otteniamo la somma per ciclo su tutto il dataframe della tecnologia
        total_bases_per_cycle = (
            tech_df.groupby("Cycle")[["match", "mismatch", "insertion"]]
            .sum()
            .sum(axis=1)   # somma match+mismatch+insertion
        )

        # Aggiunge la frazione normalizzata (%) rispetto al totale di quel ciclo
        sub["num_bases_norm"] = (
            sub.apply(lambda r: (r["num_bases"] / total_bases_per_cycle.loc[r["Cycle"]]) * 100, axis=1)
        )

        color = colori_matplotlib[j % len(colori_matplotlib)]

        # --- Riga superiore: Qscore (identità per ciclo) ---
        ax_q.plot(
            sub["Cycle"],
            qscore,
            label=tech_name,
            color=color,
            linewidth=0.8,
            marker='o',
            markersize=2,
            alpha=0.8
        )

        # --- Riga inferiore: frazione di basi normalizzata per ciclo ---
        ax_b.plot(
            sub["Cycle"],
            sub["num_bases_norm"],
            color=color,
            linewidth=0.8,
            marker='^',
            markersize=3,
            alpha=0.6
        )

    # --- Etichette e titoli ---
    ax_q.set_title(f"Sequencer Qscore {quality} bin")
    ax_q.set_ylabel("Identity-based Qscore")
    ax_b.set_xlabel("Cycle")
    ax_b.set_ylabel("Fraction of bases per cycle (%)")
    ax_q.grid(True, alpha=0.3)
    ax_b.grid(True, alpha=0.3)
    ax_q.set_ylim(0, 43)

# --- Legenda globale ---
lines, labels = [], []
for ax_i in axes_qscore:
    line, label = ax_i.get_legend_handles_labels()
    lines.extend(line)
    labels.extend(label)

fig.legend(
    lines[:len(tech_dfs)],
    labels[:len(tech_dfs)],
    loc="lower center",
    ncol=len(tech_dfs),
    fontsize=10,
    frameon=True,
)

plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```

```python
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np

results = {}

for tech_name, tech_df in tech_dfs.items():

    df = tech_df.copy()
    df["Q"] = -10 * np.log10(1 - df["identity_with_ins_filtered"])
    df = df.replace([np.inf, -np.inf], np.nan).dropna(subset=["Q"])

    # Modello con interazione
    model = smf.ols("Q ~ Cycle * C(BQ_bin)", data=df).fit()

    # Trova i bin
    bins = sorted(df["BQ_bin"].unique(), key=str)
    ref_bin = bins[0]

    tech_slopes = {}
    tech_pvalues = {}

    # --- Slope + pvalue per il bin di riferimento ---
    base_slope = model.params["Cycle"]
    base_pvalue = model.pvalues["Cycle"]

    tech_slopes[ref_bin] = base_slope
    tech_pvalues[ref_bin] = base_pvalue

    # --- Slopes e pvalue per gli altri bin ---
    for b in bins[1:]:
        term = f"Cycle:C(BQ_bin)[T.{b}]"

        slope = base_slope + model.params.get(term, 0)
        pval = model.pvalues.get(term, np.nan)

        tech_slopes[b] = slope
        tech_pvalues[b] = pval

    # Interpretazione qualitativa
    interpretation = {
        b: (
            "Crescente" if s > 0.01
            else "Decrescente" if s < -0.01
            else "Stabile"
        )
        for b, s in tech_slopes.items()
    }

    # Costruzione tabella
    results[tech_name] = pd.DataFrame({
        "Slope": tech_slopes,
        "p-value": tech_pvalues,
        "Interpretazione": interpretation
    })

# --- Output ---
for tech, df_res in results.items():
    print("\n###", tech)
    print(df_res)
```
