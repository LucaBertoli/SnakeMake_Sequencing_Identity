```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/comparazione_kit/calcolo_identit_sequenze_Sequenziatori/Element_WGS/RUN_2/IDENTITY_Element_WGS_GIAB_trimmed_basequal_strat.by_bin_cycle.tsv", delimiter="\t")

fig = plt.figure(figsize=(10,10))
colors = plt.cm.tab20.colors
for i, quality in enumerate(sorted(df["BQ_bin"].unique(), key=lambda x: str(x))):
    sub = df[df["BQ_bin"] == quality].sort_values("Cycle")
    plt.plot(sub["Cycle"], sub["identity_with_ins_filtered"], label=f"BQ {quality}", color=colors[i % len(colors)])
plt.xlabel("Cycle")
plt.ylabel("identity_with_ins_filtered")
plt.legend(title="BQ_bin")
plt.grid(True)
plt.show()
```
